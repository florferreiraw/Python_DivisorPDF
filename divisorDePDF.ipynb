```python
import tkinter as tk
from tkinter import filedialog
import PyPDF2

def dividir_PDF(arquivoEntrada, pasta_saida, numero_pagina, nome_arquivo_saida):
    with open(arquivoEntrada, 'rb') as arquivo:
        leitorPDF = PyPDF2.PdfReader(arquivo)
        
        if 1 <= numero_pagina <= len(leitorPDF.pages):
            pdf_writer = PyPDF2.PdfFileWriter()
            pdf_writer.addPage(leitorPDF.pages[numero_pagina - 1])
            
            output_file_path = f"{pasta_saida}/{nome_arquivo_saida}.pdf"
            
            with open(output_file_path, 'wb') as output_file:
                pdf_writer.write(output_file)
        else:
            print(f"Número de página inválido. O PDF possui {len(leitorPDF.pages)} páginas.")

def selecionar_arquivo():
    arquivo = filedialog.askopenfilename(title="Selecione o arquivo PDF")
    entrada_var.set(arquivo)

def selecionar_pasta_saida():
    pasta = filedialog.askdirectory(title="Selecione a pasta de saída")
    saida_var.set(pasta)

def extrair_pagina():
    arquivoEntrada = entrada_var.get()
    pasta_saida = saida_var.get()
    numero_pagina = int(pagina_var.get())
    nome_arquivo_saida = nome_saida_var.get()
    
    dividir_PDF(arquivoEntrada, pasta_saida, numero_pagina, nome_arquivo_saida)

app = tk.Tk()
app.title("Divisor de PDF")

entrada_var = tk.StringVar()
saida_var = tk.StringVar()
pagina_var = tk.StringVar()
nome_saida_var = tk.StringVar()

tk.Label(app, text="Arquivo de Entrada:").grid(row=0, column=0, sticky="w")
tk.Entry(app, textvariable=entrada_var, width=40).grid(row=0, column=1)
tk.Button(app, text="Selecionar", command=selecionar_arquivo).grid(row=0, column=2)

tk.Label(app, text="Pasta de Saída:").grid(row=1, column=0, sticky="w")
tk.Entry(app, textvariable=saida_var, width=40).grid(row=1, column=1)
tk.Button(app, text="Selecionar", command=selecionar_pasta_saida).grid(row=1, column=2)

tk.Label(app, text="Número da Página:").grid(row=2, column=0, sticky="w")
tk.Entry(app, textvariable=pagina_var, width=5).grid(row=2, column=1)

tk.Label(app, text="Nome do Arquivo de Saída:").grid(row=3, column=0, sticky="w")
tk.Entry(app, textvariable=nome_saida_var, width=20).grid(row=3, column=1)

tk.Button(app, text="Extrair Página", command=extrair_pagina).grid(row=4, column=0, columnspan=3, pady=10)

app.mainloop()
```
